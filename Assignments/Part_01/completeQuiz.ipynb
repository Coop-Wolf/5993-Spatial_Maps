```python
D = {
    "apple":"10",
    "banana":"11",
    "cherry":"8"
}

# Print keys only
for key in D.keys():
    print(key)
    
print()
    
# Print values only
for val in D.values():
    print(val)
    
print()
    
# Print values only by .get() method
for key in D.keys():
    print(D.get(key))
    
print()
    
# Print key-value
for key,val in D.items():
    print(key,  "->", val)
    
```

```python
# Cleaning

import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", None],
    "Age": ["25", "thirty", 35, np.nan, "40"],
    "Signup Date": ["2022-01-01", "not a date", "2022/03/01", None, "April 5, 2022"],
    "Score": [95.5, None, 88.0, 92.5, ""]
}

df = pd.DataFrame(data)
df
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.fillna({
    "Name": "Unknown",
    "Age": -1,
    "Signup Date": "1970-01-01",
    "Score": 0.0
})
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Score"] = pd.to_numeric(df["Score"], errors="coerce")
df["Signup Date"] = pd.to_datetime(df["Signup Date"], errors="coerce")
df.dtypes
```

```python
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Score"].fillna(df["Score"].mean(), inplace=True)
df["Signup Date"].fillna(df["Signup Date"].min(), inplace=True)
df["Name"].fillna("Unknown", inplace=True)

df
```

```python
data = {
    "first_name": ["Alice", "Bob", "Charlie", "David"],
    "last_name": ["Smith", "Jones", "Brown", "Wilson"],
    "math_score": [85, 90, 78, 92],
    "science_score": [88, 85, 82, 95]
}

df = pd.DataFrame(data)

# Creating new columns
df["average_score"] = (df["math_score"] + df["science_score"]) / 2

# Renaming column header
df.rename(columns={"math_score": "Math", "science_score": "Science"}, inplace=True)

# Functino to calculate grade
def grade(score):
    if score >= 90:
        return "A"
    elif score >=80:
        return "B"
    elif score >=70:
        return "C" 
    elif score >= 60:
        return "D"

# Creating new column for grades
df["Grade"] = df["average_score"].apply(grade)

df
```

```python
# Adding some columns for more practice
df["Bonus"] = 5
df["average_score"] = df["average_score"] + df["Bonus"]
df["Grade"] = df["average_score"].apply(grade)
df
```
